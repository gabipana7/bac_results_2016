```python
import pandas as pd
import re
```

```python
with open("data/bac_2015.csv", "r", encoding="utf-16") as file:
    content = file.read()

# Replace commas in numbers with a regular expression

processed_content = re.sub(r'(\d+),(\d+)', r'\1.\2', content)

# Write the processed content to a new file
with open("data/bac_2015_processed.csv", "w", encoding="utf-16") as file:
    file.write(processed_content)
```

```python
df = pd.read_csv("data/bac_2015_processed.csv", sep='\t', encoding = "utf-16")
```

```python
df['NOTA_EA'] = pd.to_numeric(df['NOTA_EA'], errors='coerce')
df['NOTA_EB'] = pd.to_numeric(df['NOTA_EB'], errors='coerce')
df['NOTA_EC'] = pd.to_numeric(df['NOTA_EC'], errors='coerce')
df['NOTA_ED'] = pd.to_numeric(df['NOTA_ED'], errors='coerce')
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import percentileofscore
```

```python
# Dana's results
# my results
record = df[(df['NOTA_EA'] == 9.70) & (df['NOTA_EC'] == 10.0) & (df['NOTA_ED'] == 9.85)]
# cod candidat: 252299
# Unitate: 2061100032
```

```python
df = df[(df[['NOTA_EA', 'NOTA_EC', 'NOTA_ED', 'Medie']] <= 10.0).all(axis=1)]
```

```python
my_scores = {'NOTA_EA': 9.7, 'NOTA_EC': 10.0, 'NOTA_ED': 9.85, 'Medie': 9.85}
```

```python
# Set the size of the plots
plt.figure(figsize=(12, 8))

# Plot distributions and highlight the student's score
for i, subject in enumerate(['NOTA_EA', 'NOTA_EC', 'NOTA_ED', 'Medie']):
    plt.subplot(2, 2, i+1)  # 2 rows, 2 columns, i+1th subplot
    sns.histplot(df[subject], kde=True, color='blue', bins=10, label=subject)
    
    # Calculate the percentile of the student's score
    percentile = percentileofscore(df[subject], my_scores[subject])
    
    # Add a vertical line at the student's score
    plt.axvline(my_scores[subject], color='red', linestyle='--', label=f"Student {subject} Score")
    
    # Add text annotation with the percentile
    plt.text(my_scores[subject] + 0.5, 2, f"{percentile:.1f}%", color='red', fontsize=12)
    
    plt.title(f'Distribution of {subject} Scores')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.legend()

# Adjust layout to prevent overlap
plt.tight_layout()
plt.savefig('results/2015/nationwide.png')
plt.show()
```
